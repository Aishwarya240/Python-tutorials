guess character game


```python
import random 

word_list = ["aardvark", "baboon", "camel"]

chosen_word = random.choice(word_list)

guess = input("Guess a letter: ").lower()

n=0
while n <= len(chosen_word)-1:
    n += 1
    if guess == chosen_word[n-1]:
        print("Right")
    else:
        print("Wrong")
```

```python
import random 

word_list = ["aardvark", "baboon", "camel"]

chosen_word = random.choice(word_list)

guess = input("Guess a letter:").lower()

for letter in chosen_word:
    if guess == letter:
        print("Right")
    else:
        print("Wrong")

```

```python
#key word argument

def my_function(name, location):
    print(f"Hello {name}")
    print(f"Your location is {location}")
    
my_function(location="Pune", name="Sakib")
```

```python
#program to get number of can required to paint a given height and width wall.

def paint_cal(height, width, cover):
    num_of_can = (height * width) / cover
    print(f"number of can required: {round(num_of_can)}")
    
test_h = int(input("Enter Height of Wall: "))
test_w = int(input("Enter Width of Wall: "))
coverage = 5
    
paint_cal(height=test_h, width=test_w, cover=coverage)
```
